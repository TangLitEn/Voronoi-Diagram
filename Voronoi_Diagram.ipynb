```python
############################################
## DO NOT MODIFY THIS PORTION OF CODE!!!! ##
############################################
def create_zero_matrix(n,m):
    return [[0 for i in range(m)] for j in range(n)]

def m_tight_print(m):
    for i in range(len(m)):
        line = ''
        for j in range(len(m[0])):
            line += str(m[i][j])
        print(line)

############################################
########### End of do not modify ###########
############################################

############
##  Task  ##
############

def calc_distance(x1,y1,x2,y2):
    return abs(x1-x2)**2 + abs(y1-y2)**2

# don't need to square to satisfied the integer arithmetic
# original pythagoras theorem is sqrt(abs(x1-x2)**2 + abs(y1-y2)**2)
# let's say we have 3 distance,A,B,C, where the order is A > B > C
# A^2 > B^2 > C^2 , the order remain the same even after squaring
# we can modified the pythagoras theorem to abs(x1-x2)**2 + abs(y1-y2)**2

def closest_site(x,y,sites):
    distance = []
    for site_index in range(len(sites)):
      distance.append(calc_distance(x,y,sites[site_index][0],sites[site_index][1]))
    min_distance = min(distance)
    if(distance.count(min_distance)>1):
      return "X"
    else:
      return distance.index(min_distance)

def pd_map(r,c,sites):
    matrix = create_zero_matrix(r,c);
    for row in range(r):
      for column in range(c):
        matrix[row][column] = closest_site(row,column,sites)
    return matrix

print("")
m_tight_print(pd_map(50,80,[[20,10], [30,30],[40,20],[45,55],[10,55],[35,70],[35,60]]))
print("")
m_tight_print(pd_map(10,10,[[2,3],[4,9],[7,2]]))
print("")
m_tight_print(pd_map(60,70,[[10,20],[30,20],[40,50]]))
print("")

# ex = pd_map(60,70,[[10,20],[30,20],[40,50]])
```

```python
# additional question
# for aditional task
from PIL import Image
import random
import string

# Define a color mapping for the values in the matrix
color_mapping = {
    "X": (255, 255, 255),  # White color for "X"
}

def update_colour_mapping(sites):
    for site_index in range(len(sites)):
      color_mapping[site_index] = (random.randint(0,256),random.randint(0,256),random.randint(0,256))

def create_colored_image(matrix, color_mapping):
    height, width = len(matrix), len(matrix[0])
    image = Image.new('RGB', (width, height))

    for y in range(height):
        for x in range(width):
            value = matrix[y][x]
            color = color_mapping.get(value, (0, 0, 0))
            image.putpixel((x, y), color)
    image.show()
    return image

def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    return result_str

def pd_map_2(r,c,sites):
    matrix = create_zero_matrix(r,c);
    for row in range(r):
      for column in range(c):
        matrix[row][column] = closest_site(row,column,sites)
    update_colour_mapping(sites)
    image = create_colored_image(matrix, color_mapping)
    name = get_random_string(8) + ".png"
    image.save(name)
    print("image save as: ",name)
    return matrix

print("")
m_tight_print(pd_map_2(50,80,[[20,10], [30,30],[40,20],[45,55],[10,55],[35,70],[35,60]]))
print("")
m_tight_print(pd_map_2(10,10,[[2,3],[4,9],[7,2]]))
print("")
m_tight_print(pd_map_2(60,70,[[10,20],[30,20],[40,50]]))
print("")
```
